```python
import pandas as pd
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")
sub = pd.read_csv("../exp_13/output/submit.csv")
```

```python
from collections import Counter
import numpy as np
```

```python
test = test.sort_values("ncode").reset_index(drop=True)
```

```python
cnt = 0
ncode_dict = {}
for i, row in test.iterrows():
    if row["writer"] in list(train["writer"]):
        l = [0]*5
        d = {0:0, 1:0, 2:0, 3:0, 4:0} 
        tmp_list = train.query("writer == @row['writer']")["fav_novel_cnt_bin"].to_list()
        for j in tmp_list:
            d[j] += 1
        ncode_dict[row["ncode"]] = list(d.values()) / np.sum(list(d.values()))
        cnt+=1
    else:
        ncode_dict[row["ncode"]] = np.nan

print(cnt)
```

```python
cnt = 0
for i in ncode_dict.keys():
    if type(ncode_dict[i]) == float:
        cnt += 1 
        continue
    else:
        sub.loc[cnt, 1:] = ncode_dict[i]
        cnt += 1
```

```python
sub.to_csv("../output/submit_rulebase.csv", index=False)
```
